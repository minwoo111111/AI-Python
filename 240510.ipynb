```python
#데이터프레임에서 행과 칼럼을 지정하여 데이터 보기
#데이터프레임명.loc[행레이블, 열레이블]
#데이터프레임명.iloc[행위치,열위치]

data_loc=df7.loc[2,['반']]
data_iloc=df7.iloc[2,3]
print(data_loc)
print(data_iloc)
```

```python
#데이터프레임에 대한 조건 연산 결과 출력하기

#df7에서 국어점수가 80점 이상인 데이터프레임을 만들려면...
condition2=df7['국어']>80
print(condition2)   #이상태에선 True False값으로만 나옴. 연산결과만 나오는 거임.

condition3=df7[df7['국어']>80]  #이렇게 해주면 됨
print(condition3)

#국어가 80점 이상이고 수학이 90점 이상인 사람만
condition4=df7[(df7['국어']>80) & (df7['수학']>90)]
print(condition4)
```

```python
#데이터프레임의 데이터 정렬
#데이터프레임명.sort_values(by=['칼럼명'], ascending=True/False, inplace=True/False)
            #이때 ascending은 오름차순으로 할지 말지 결정하는것. ascending 안 써주면 걍 오름차순으로 정렬됨.
            #inplace는 원본순서를 바꿀지말지를 지정. 기본은 false임. 굳이 안 쓰는 게 좋다.

#국어점수가 높은 사람부터 표현하라
a=df7.sort_values(by=['국어'], ascending=False)    #이렇게하면 됨..
print(a)
```

```python
#데이터프레임의 속성값 정보 보기
#.index -->데이터프레임이 갖는 행번호
#.columns -->데이터프레임이 갖는 칼럼명들
#.size -->데이터프레임에 있는 개별 항목 개수들 (즉, 행의개수*칼럼개수. 총요소의 수)
```

```python
#데이터프레임의 전체 기본 통계 출력
#데이터프레임명.describe()

#칼럼값별 데이터개수 보기
#데이터프레임명['칼럼이름'].value_counts()
#남자 몇명이고 여자 몇명인지 알아보시오
b=df7['남'].value_counts()
print(b)
```

```python
#개별칼럼의 통계값 구하기.
#sum(), count(), mean(), std(), max(), min(), median 등 통계함수로 원하는 통계값 계산 가능

#그룹별 통계값 구하기
#데이터프레임명.groupby('칼럼명').통계함수명()    groupby()로 그룹으로 묶을 칼럼을 지정한 뒤 sum(),count(),mean()등 통계함수를이용해 칼럼별로 구분한 통계값 계싼

#데이터프레임의 결측치처리
#데이터프레임에서 결측치확인: 데이터프레임명.isna() --> 칼럼의값이없으면 true, 있으면 false
                              데이터프레임명.isna().sum() --> 각 칼럼별 결측치의 개수 집계
#데이터프레임에서 결측치에 값 채우기: 데이터프레임명.fillna(채울값, inplace=True/False)
                                      데이터프레임명.fillna({'칼럼명':'채울값'}, inplace=True/False) -->칼럼에 있는 결측치를 다른 값으로 채운다.
#데이터프레임에서 결측치가 있는 데이터 삭제: 데이터프레임명.dropna(axis=0/1, subset=['칼럼명'], inplace=True/False)  -->그칼럼에서 결측치가있는 레코드를 날려버림
```

```python
#데이터프레임의 칼럼을 이용하여 새 칼럼 추가
#데이터프레임의 '국어'와 '수학' 칼럼값을 더해 '총점'이라는 새로운 칼럼 생성하기
df7['총점']=df7['국어']+df7['수학']

#데이터프레임의 행/칼럼 삭제
#데이터프레임명.drop(인덱스값/칼럼명, axis=0/1, inplace=False/True) --> 인덱스값으로 지정한 레코드/칼럼명으로 지정한 칼럼을 데이터프레임에서 삭제

#데이터프레임의 항목값 일괄 변경
#데이터프레임명.replace(바꿔야하는값, 바꿀값, inplace=False/True)
#데이터프레임명.replace({바꿔야 하는 값1:바꿀 값1, … }, inplace=False/True )
#데이터프레임명.replace({‘칼럼명’: {바꿔야 하는 값1:바꿀 값1, … } }, inplace=False/True )  <--데이터프레임에서 바꿔야 하는 값을 바꿀 값으로 일괄 변경

```
