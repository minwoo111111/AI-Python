```python
#예제3: 두 개의 그래프 표현하기 <--데이터 프레임의 두 칼럼에 대해 이차원 그래프 만들기!

df=pd.read_excel('data1.xlsx')

plt.plot(df['kor'])
plt.plot(df['math'])


plt.show()

###이때, 두 그래프가 헷갈리지 않게 다르게 시각화하려면, 선스타일과 마커 지정!
plt.plot(df['kor'], '--', marker='o')
plt.plot(df['math'], marker='^')

plt.show()


####그래프에 범례 표시를 하고 싶으면,
    #1. 데이터에 레이블을 각각 지정해준 후
    #2. plt.legend()를 사용

plt.plot(df['kor'], '--', marker='o',label='korean')
plt.plot(df['math'], marker='^',label='math')
plt.legend() #<--ㅡ우리가 붙인 레이블로 범례를 표시해라
plt.show
```

#실습문제
###다음은 한 학생의 일주일 간의 공부 시간과 성적 데이터입니다.
'''
요일, 공부시간(시간), 성적
월, 2, 70
화, 3, 75
수, 1, 65
목, 5, 85
금, 4, 80
토, 2, 70
일, 1, 60
'''

1. 왼쪽 데이터를 이용하여 데이터프레임을 생성하시오. (pd.dataframe() 활용하여)
2. 생성한 데이터 프레임을 이용하여 요일별 공부시간을 선 그래프로 나타내시오. x축은 요일, y축은 공부시간임.
3. 생성한 데이터프레임을 이용하여 요일별 성적을 선 그래프로 나타내시오. x축은 요일, y축은 성적임.


```python
#데이터프레임을 생성하기
import pandas as pd
df=pd.DataFrame({'요일':['월','화','수','목','금','토','일'],
                '공부시간':[2,3,1,5,4,2,1],
                '성적':[70,75,65,85,80,70,60]})

#생성한 데이터프레임을 이용해 요일별 공부시간을 선 그래프로 나타내기
x=df['요일']
y=df['공부시간']
plt.plot(x,y)
plt.show()

#생성한 데이터프레임을 이용해 요일별 성적을 선 그래프로 나타내기.
a=df['요일']
b=df['성적']
plt.plot(a,b)
plt.show()

#둘이 같이 출력하면 한눈에 비교할 수 있을 것
plt.plot(x,y)
plt.plot(a,b)
plt.show()
```

'''

'''

1.왼쪽 데이터를 이용하여 도시별 인구 수를 수직 막대 그래프로 시각화하라
2. 그래프의 제목을 도시별 인구 수로 설정하라
3. x축 라벨을 '도시'로, 'y축 라벨을 '인구 수'로 설정하라.

1. penguins 데이터를 가져온 후 펭귄 품종(species)별 몸무게(body_mass_g)의 평균을 그래프로 그리는데,아래 그림과 같이 성별을 구분해서 그려보자.
2. penguins 데이터의 숫자값을 갖는 항목들끼리의 상관관계 그래프를 그려보자.
3. 2번 미션의 결과에서 칼럼들간의 상관관계를 확인하고 관계가 있는 두 칼럼을 선택하여 산점도와 추세선
이 함께 나타나는 그래프를 그려보자.


```python
#1.
penguindata=pd.read_csv('penguins.csv')
sns.barplot(x='species',y='body_mass_g',data=penguindata, hue='sex')   #이때 data가 뭔지 지정해줘야 x='species'라고 했을 때 알아들음
                                                          #이때 hue='sex'라고 해주면 성별에따라 구분돼서 몸무게평균이 나옴
plt.show()

#2.
sns.pairplot(penguindata, hue='species')   #hue가 species이므로 종별로 데이터의 숫자값 갖는 칼럼들끼리의 상관관계를 보여줌
plt.show()

#3.
sns.regplot(x='flipper_length_mm', y='body_mass_g', data=penguindata)
plt.show()
```

#숫자값을 갖는 칼럼들의 상관관계를 보여주는 그래프: pairplot(데이터프레임, hue=‘범주를 갖는 칼럼명＇)


```python
iris_data=pd.read_csv('iris.csv')
sns.pairplot(iris_data,hue='species')
plt.show()
```

#산점도와 추세선(회귀선)을 동시에 보여주는 그래프 : regplot( )
•regplot(x=‘x축에 표시할 칼럼명‘, y=‘y축에 표시할 칼럼명‘, 데이터프레임)

```python
iris_data=pd.read_csv('iris.csv')
sns.regplot(x='petal_length',y='petal_width', data=iris_data)
plt.show()
```

#집계한 값에 따라 색깔을 표시하는 그래프 : heatmap( )
heatmap(data=2차원 행렬 형태로 전달된 데이터, annot=True/False,fmt=‘파이썬 서식 문자열‘, cbar=True/False)

색이 진할수록 상관관계가 높다는 것.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

iris_data=pd.read_csv('iris.csv')
df=iris_data.corr()
sns.heatmap(data=df,annot=True,fmt='.2f',cbar=False)
plt.show()
```
